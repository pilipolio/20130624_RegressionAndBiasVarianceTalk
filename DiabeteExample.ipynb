## Diabetes Example

[Data set](http://cran.r-project.org/web/packages/lars/index.html) from the lars paper.

```python
!wget http://cran.r-project.org/src/contrib/lars_1.2.tar.gz
!tar -zxvf lars_1.2.tar.gz lars/data/diabetes.RData
```

```python
%load_ext rmagic 
%R load('lars/data/diabetes.RData')

diabetes_x = %Rget diabetes$x
diabetes_x2 = %Rget diabetes$x2
diabetes_y = %Rget diabetes$y

from sklearn.cross_validation import train_test_split
X_train, X_test, X2_train, X2_test, y_train, y_test = train_test_split(diabetes_x, diabetes_x2, diabetes_y, train_size=.7)
```

### $L_1$ regularisation with the quadratic model

```python
from sklearn.linear_model import LassoCV
lasso_cv = LassoCV(alphas=np.logspace(-3, 1, 100), cv=10, max_iter=10000)
lasso_regressor = lasso_cv.fit(X2_train, y_train)

print "LASSOCV chose a lambda {:.4f} with {} out of {} variables and yields TRAIN/TEST R square = {:.2f}/{:.2f}".format(
    lasso_regressor.alpha_,
    np.sum(lasso_regressor.coef_!=0),
    len(lasso_regressor.coef_), 
    lasso_regressor.score(X=X2_train, y=y_train), lasso_regressor.score(X=X2_test, y=y_test))
```

```python
plt.plot(-np.log10(lasso_regressor.alphas_),
    np.transpose(np.mean(lasso_regressor.mse_path_, axis=1)), 'k-'
);
y_lims = plt.ylim()
plt.vlines(-np.log10(lasso_regressor.alpha_), *y_lims, color='grey', linestyles='--')
plt.ylim(y_lims);
plt.xlim([-.5, 3]);
```

```python
# https://github.com/scikit-learn/scikit-learn/blob/master/examples/linear_model/plot_lasso_coordinate_descent_path.py
f, ax = plt.subplots(1,1)
f.set_size_inches(18, 9)

l1 = plt.plot(-np.log10(lasso_regressor.alphas_), lasso_regressor.coef_path_, color='gray', alpha=.6)

mask = lasso_regressor.alphas_ >= lasso_regressor.alpha_

l1 = plt.plot(-np.log10(lasso_regressor.alphas_[mask]), lasso_regressor.coef_path_[mask,:], alpha=.8)

plt.vlines(x=-np.log10(lasso_regressor.alpha_), ymin=-800, ymax=800)

plt.xlabel('$- log (\lambda)$')
plt.ylabel('coefficients')
plt.title('Lasso Path')
#plt.legend(l1[-1], 'Lasso', loc='lower left')
plt.axis('tight');
```
