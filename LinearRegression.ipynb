## Linear regression and MLE

Model : $$ y = \mathbf{x}^T \mathbf{w} + \epsilon $$ where $\mathbf{x}$ and $\mathbf{w} \in \mathbb{R}^d$
    
The ordinary least square estimation of $\mathbf{w}$ :
            
$$ \hat{\mathbf{w}}_{ols} = \underset{\mathbf{w}}{\text{arg min}} \ || \mathbf{x}^T \mathbf{w} - y||_2^2 $$

With a data-set $\mathcal{D}$ made of $n$ observations : $\mathcal{D} = (\mathbf{x}_i, \mathbf{y}_i) _ {i = 1, \ldots, n}$

 - $ \mathbf{X} = (\mathbf{x}_1, \ldots, \mathbf{x}_n)^T \in \mathcal{M} _ {n,d}$ 

 - $ \mathbf{y} = (\mathbf{y}_1, \ldots, \mathbf{y}_n)^T \in \mathcal{M} _ {n,1}$

$$ \mathbf{X} \hat{\mathbf{w}} _ {ols} = \mathbf{y} \Rightarrow \hat{\mathbf{w}} _ {ols} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y} $$

When $ \epsilon \sim \mathcal{N}(0, \sigma^2) $, Maximum-Likelihood-Estimator $\hat{\mathbf{w}} _ {ols} = \hat{\mathbf{w}} _ {mle}$

## Toy example with a known function $f$

$f$ polynom of degree $d = 3$ (with null intercept $w_0 = 0$).

$ f : x \mapsto  w_1 \ x + w_2 \ x^2 + w_3 \ x^3$

$n$ values $\mathbf{x}_i \in \mathbb{R}^d \ $ drawn from $X \sim \mathcal{U}(0,1) $

Polynomial function basis :

- $\mathbf{X}_{\bullet,1} = (x_0, \ldots, x_n)$
- $\mathbf{X}_{\bullet,2} = (x_0^2, \ldots, x_n^2)$
- $\mathbf{X}_{\bullet,3} = (x_0^3, \ldots, x_n^3)$


```python
n = 25
d = 3
# No intercept
w_oracle = [.25, -.25, -.1]

from numpy.polynomial.polynomial import Polynomial
def create_polynom(w, intercept=0, domain=[-1, 1]):
    """ With default null intercept"""
    return Polynomial([intercept] + list(w), domain=domain)

x_n = np.sort(np.random.uniform(low=-1, high=1, size=n))
f_oracle = create_polynom(w_oracle)

print 'Oracle coefficients = {} and ||w||_2 = {}'.format(w_oracle, np.linalg.norm(w_oracle))
```

### MLE solution:

$$ \hat{\mathbf{w}}_{mle} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

```python
y_n = f_oracle(x_n) + np.random.normal(scale=.1, size=n)
X_n_d = np.vstack((x_n, np.power(x_n,2), np.power(x_n,3))).T
w_ols = np.dot(np.dot(np.linalg.inv(np.dot(X_n_d.T, X_n_d)), X_n_d.T), y_n)

print 'OLS coefficients = {} and ||w||_2 = {}'.format(w_ols, np.linalg.norm(w_ols))

def plot_polynomial_fit(oracle_P, ols_fit_P, fit_label, fit_alpha=.75, fig=None, ax=None):
    if type(oracle_P) is not Polynomial:
        oracle_P = create_polynom(oracle_P)
    if type(ols_fit_P) is not Polynomial:
        ols_fit_P = create_polynom(ols_fit_P)
    if fig is None:
        f, ax = plt.subplots(1,1, figsize=(8, 6))
    ax.plot(*oracle_P.linspace(), color="b", ls="-", lw=3, alpha=.7, label="$y=f(x)$")
    ax.plot(x_n, y_n, "ko", alpha=.7,label="$(y_i)_{i=1,\ldots,n}$")
    ax.plot(*ols_fit_P.linspace(), color="g", ls="-", lw=2, alpha=fit_alpha, label=fit_label)
    plt.legend(loc="lower right", prop={'size':20})
    return fig, ax

fig, ax = plot_polynomial_fit(w_oracle, w_ols, "$y=\hat{f}_{ols}(x)$")
```

### Coordinates descent algorithms

Iteratively build an estimator $\hat{\mu} _ \mathcal{A}$ by progressivly adding the _most influential inputs_ into a subset $\mathcal{A}$ of the indices $\{ 1, 2, \ldots, d \}$ :
    
 - At step $i$, estimator $\mu _ \mathcal{A}$ and residuals $\mathbf{y} - \mu  _ \mathcal{A}$
    
 - Update $\mathcal{A} \rightarrow \mathcal{A} + j$
    
 - Update $\hat{\mu} _ {\mathcal{A}} \rightarrow \hat{\mu} _ {\mathcal{A}} + \gamma \mathbf{X} _ {\mathcal{A}} $
where     
        $$ \mathbf{X} _ {\mathcal{A}}  = (\cdots, \mathbf{X} _ {\bullet,j}, \cdots ) _ {j \in \mathcal{A}}$$

<img src="https://raw.github.com/pilipolio/20130624_RegressionAndBiasVarianceTalk/master/linear_regression_geometrical_illustration.png" alt="Drawing" style="width: 800px;"/>

#### Forward selection or forward stepwise regression

- Classic model selection method : [Weisberg (1980)](http://www.stat.umn.edu/alr/preface.html)

- Assume $\mathbf{x}_j$ with variances equal to 1

- Choose $\hat{j} = \underset{j}{\text{arg max}} \ | \hat{c_j} | $ where $\hat{c} = \mathbf{X}^T (\mathbf{y} - \mu)$

- $\hat{\mu} \rightarrow \hat{\mu} + \gamma \ \text{sign} (\hat{c _ j}) \ \mathbf{x}  _  \hat{j}  $  with $ \gamma = | \hat{c  _  \hat{j}} | $

#### Forward _stagewise_ regression

- More cautious version of the forward selection
- Same direction $ \text{sign} (\hat{c _ j}) \ \mathbf{x}  _  \hat{j}  $, but with an infinitesimal step $\gamma > 0$

```python
##################
### Initialisation
##################
i = 0
# no intercept in x and mu because w_0 = 0
X_n_d = np.vstack((x_n, np.power(x_n,2), np.power(x_n,3))).T

m = 1000
w_m_d = np.zeros((m, d))
mu_d = np.dot(X_n_d, w_m_d[i,:])
mu_d
```

```python
### Step i -> i+1
eps = 1
## Correlation need to be normalised by each x variance
hat_c_d = np.dot(X_n_d.T, y_n - np.dot(X_n_d, w_m_d[i,:])) / [np.dot(x, x) for x in X_n_d.T]
hat_j = np.argmax(np.abs(hat_c_d))

print 'At step {}, hat_c_d = {} and hat_j = {}'.format(i, hat_c_d, hat_j)
w_m_d[i+1, :] = w_m_d[i, :]
w_m_d[i+1, hat_j] = w_m_d[i+1, hat_j] + eps * hat_c_d[hat_j]

print 'w_d ({})= w_d ({}) + hat_c_d[{}] ({})'.format(w_m_d[i+1,:], w_m_d[i,:], hat_j, w_m_d[i+1,:] - w_m_d[i,:])

i += 1

fig, ax = plot_polynomial_fit(w_oracle, w_m_d[i,:], fit_alpha=.75, fit_label="$y=\hat{f}_{stepwise}(x)$")
```

### LARS algorithm:

- Middle ground between the greedy forward selection and the slow stagewise.
- Analytically select direction and amplitude to descent along _equiangular vectors_

![caption](https://raw.github.com/pilipolio/20130624_RegressionAndBiasVarianceTalk/master/these_lars_page.png)

- The algebra is unfortunately much more complex than the geometrical formulation...
- Note $ \mathbf{u} _ { \mathcal{A} }$ the equi-angular vector

$$ \mathbf{u} _ { \mathcal{A} } \propto \mathbf{X} _ { \mathcal{A} } (\mathbf{X}^T _ { \mathcal{A} } \mathbf{X} _ { \mathcal{A} } ) ^ { -1 } \mathbf{1} _ { \mathcal{A} } $$

- $ \mathbf{u}  _ { \mathcal{A} }$ make equal angles with the column of $\mathbf{X} _ { \mathcal{A} }$

- The LARS step $\gamma$ is chosen so that $\forall j \in \mathcal{A},$

$$ | c_j (\gamma) | = \hat{C} - \gamma \text{ constant} $$

- And so that $\forall k \in \mathcal{A} ^ { c }$

$$ | c_k (\gamma) | \leq  \underset{j} { \text{ max }} | c_j (\gamma) | $$

$$ | c_j (\gamma) | = \hat{C} - \gamma $$

- $\hat{\mu} _ { \mathcal{A} } \rightarrow \hat{\mu} _ { \mathcal{A} } + \gamma \ \mathbf{u}  _ { \mathcal{A} }$


### Penalized or regularised estimators :

$$ \underset{\mathbf{w}}{\text{arg min}} \ || \mathbf{x}^T \mathbf{w} - y||_2^2 + \underbrace{ \lambda || \mathbf{w} ||_q } _ {penalty} $$

$q = 2 \Rightarrow$ ridge regression $\sim$ gaussian centered _a priori_ with variance $\propto \lambda^{-1}$

$$ \hat{\mathbf{w}}_{ridge} = (\mathbf{X}^T \mathbf{X} + \lambda \mathbf{I} )^{-1} \mathbf{X}^T \mathbf{y}$$

$q = 1 \Rightarrow$ lasso regression

 - $L_1$ penalty yield sparser models than the $L_2$ one

 - Initial paper from Tibshirani (quadratic optimisation)

 - Closely related to descent algorithm $\Rightarrow$ a simple constraint on the LARS enable to calculate the full path in $d$ steps





```python
def ridge_regression(X_n_d, y_n, lbda):
    d = X_n_d.shape[1]
    return np.linalg.solve(np.dot(X_n_d.T, X_n_d) + np.diag(np.repeat(lbda, d)), np.dot(X_n_d.T, y_n))

plt.plot(x_n, y_n, "ko", alpha=.7,label="$(y_i)_{i=1,\ldots,n}$")

for l, alpha in reversed(zip(np.logspace(-1, 3, 10), np.linspace(.5, 0, 10))):
    w_ridge_l = ridge_regression(X_n_d, y_n, l)
    ridge_P = create_polynom(w_ridge_l)
    plt.plot(*ridge_P.linspace(), color="purple", ls="-", lw=2, alpha=alpha)

plt.plot(*ridge_P.linspace(), color="purple", ls="-", lw=2, alpha=.75, label="$y=\hat{f}_{\lambda=0}(x)$")
plt.plot(*f_oracle.linspace(), color="b", ls="-", lw =3, alpha=.7,label="$y=f(x)$")
plt.legend(loc="lower right", prop={'size':20})
plt.figsize(8, 6)
```

### Regularisation paths


```python
# fixed w_0 and w_1 to be the OLS ones
w_1 = w_ols[0]
w2_grid, w3_grid = np.meshgrid(np.linspace(-.5, .5, num=100), np.linspace(-.5, .5, num=100))

# x, y and z 2D grids
error_grid = np.reshape(
    [np.linalg.norm(np.dot(X_n_d, [w_1, w2, w3]) - y_n) 
        for w2, w3 in zip(w2_grid.ravel(), w3_grid.ravel())], (100, 100))
penalization_grid = np.reshape(
    [np.linalg.norm([w_1, w2, w3])
        for w2, w3 in zip(w2_grid.ravel(), w3_grid.ravel())], (100, 100))

# Building regularisation paths
w_2_path, w_3_path = zip(*[ridge_regression(X_n_d, y_n, lbda)[1:] for lbda in np.logspace(-2, 3, 100)])

from sklearn.linear_model import lars_path
lasso_lambdas_m, lasso_active_m, w_lasso_d_m = lars_path(X_n_d, y_n, method='lasso')

w_lasso_2_path, w_lasso_3_path = w_lasso_d_m[1:,:]

def plot_weight_space(ax, x_grid, y_grid, z_grid):
    ax.contour(x_grid, y_grid, z_grid)
    ax.vlines(0, -.5, .5, alpha=.5)
    ax.hlines(0, -.5, .5, alpha=.5)
    ax.set_xlim(-.5, .5)
    ax.set_ylim(-.5, .5)

f, (ax1, ax2, ax3) = plt.subplots(1,3)
f.set_size_inches(18, 6)

plot_weight_space(ax1, w2_grid, w3_grid, penalization_grid)
ax1.scatter(w_oracle[1], w_oracle[2], marker="o", color="lightgrey", s=80, linewidth=3);
ax1.text(w_oracle[1], w_oracle[2], s="$w_{oracle} \ $", size=18, weight='bold', va='bottom', ha='right');

plot_weight_space(ax2, w2_grid, w3_grid, error_grid)
ax2.scatter(w_ols[1], w_ols[2], marker="o", color="lightgrey", s=80, linewidth=3)
ax2.text(w_ols[1], w_ols[2], s="$ \ \hat{w}_{ols}$", size=18, weight='bold')
ax2.scatter(w_oracle[1], w_oracle[2], marker="o", color="lightgrey", s=80, linewidth=3);
ax2.text(w_oracle[1], w_oracle[2], s="$w_{oracle} \ $", size=18, weight='bold', va='bottom', ha='right');

plot_weight_space(ax3, w2_grid, w3_grid, error_grid + penalization_grid)

ax3.scatter(0, 0, marker="o", color="lightgrey", s=80, linewidth=3)
ax3.scatter(w_oracle[1], w_oracle[2], marker="o", color="lightgrey", s=80, linewidth=3);
ax3.scatter(w_ols[1], w_ols[2], marker="o", color="lightgrey", s=80, linewidth=3)

ax3.plot(w_2_path, w_3_path, color='k', lw=2, ls='--', label='ridge path')
ax3.plot(w_lasso_2_path, w_lasso_3_path, color='grey', lw=2, ls='-', label='lasso path')
ax3.text(w_ols[1]+0.025, w_ols[2]-0.025, s="$\lambda=0$", size=14, weight='bold', ha='left', va='top')
ax3.text(0.025, 0.025, s="$\lambda=+\infty$", size=14, weight='bold')

plt.legend();
```

## Model selection and the Bias-Variance trade-off

### Bias-Variance decomposition
Mostly from that [course](http://www.youtube.com/watch?v=C3nIFH649wY&list=PLD0F06AA0D2E8FFBA&index=70)

- From a set of observed data $\mathcal{D}$, $\hat{\theta} = f(\mathcal{D})$ is an estimator of $\theta$. 
- Using the _squared-loss_ error function, the risk function $R(\hat{\theta};\theta, \mathcal{D}) = \mathbb{E}[(\hat{\theta} - \theta)^2 | \theta]$ can decomposed by :
    
$$ 
R(\hat{\theta};\theta, \mathcal{D}) = \underbrace{\mathbb{E}[(\hat{\theta} - \mathbb{E}[\hat{\theta}])^2]} _ {\text{variance}} + \underbrace{(\mathbb{E}[\hat{\theta}] - \theta)^2} _ {\text{bias}^2}
$$
