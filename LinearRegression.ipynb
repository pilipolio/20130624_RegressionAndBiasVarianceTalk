## Linear regression and MLE

Model :

$$ y = \mathbf{x}^T \mathbf{w} + \epsilon $$
    
where :
    
 - $\mathbf{x}$ and $\mathbf{w} \in \mathbb{R}^d$
 - $ \epsilon \sim \mathcal{N}(0, \sigma^2)$
    
The Maximum-Likelihood-Estimator (aka Ordinary Least Square) estimation of $\mathbf{w}$ :
            
$$ \hat{\mathbf{w}} = \underset{\mathbf{w}}{\text{arg min}} \ || \mathbf{x}^T \mathbf{w} - y||_2^2 $$

With a data-set $\mathcal{D}$ made of $n$ observations : $\mathcal{D} = (\mathbf{x}_i, \mathbf{y}_i)_{i = 1, \ldots, n}$

 - $ \mathbf{X} = (\mathbf{x}_1, \ldots, \mathbf{x}_n)^T \in \mathcal{M}_{n,d}$ 

 - $\mathbf{y} = (\mathbf{y}_1, \ldots, \mathbf{y}_n)^T \in \mathcal{M}_{n,1}$

$$ \mathbf{X} \hat{\mathbf{w}}_{ols} = \mathbf{y} \Rightarrow \hat{\mathbf{w}}_{ols} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

## Illustration with a polynom

$$ y = f(x) + \epsilon$$

Oracle solution : $f(x) = w_0 + w_1 \ \mathbf{x}_1 + w_2 \ \mathbf{x}_2 + w_3 \ \mathbf{x}_3$

$n$ values $x_i \in \mathbb{R} \ $ drawn from $X \sim \mathcal{U}(0,1) $

```python
n = 25
d = 3
w_oracle = [0, .25, -.25, -.1]

from numpy.polynomial.polynomial import Polynomial
x_n = np.sort(np.random.uniform(low=-1, high=1, size=n))
f_oracle = Polynomial(w_oracle, domain=[-1, 1])

print 'Oracle coefficients = {} and ||w||_2 = {}'.format(w_oracle, np.linalg.norm(w_oracle))
```

###OLS solution:

$$ \hat{\mathbf{w}}_{ols} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

```python
y_n = f_oracle(x_n) + np.random.normal(scale=.1, size=n)
X_n_d = np.vstack((np.repeat(1, n), x_n, np.power(x_n,2), np.power(x_n,3))).T
w_ols = np.dot(np.dot(np.linalg.inv(np.dot(X_n_d.T, X_n_d)), X_n_d.T), y_n)

print 'OLS coefficients = {} and ||w||_2 = {}'.format(w_ols, np.linalg.norm(w_ols))

def plot_polynomial_fit(oracle_P, ols_fit_P, fit_label, fit_alpha=.75, fig=None, ax=None):
    if type(oracle_P) is not Polynomial:
        oracle_P = Polynomial(oracle_P, domain=[-1, 1])
    if type(ols_fit_P) is not Polynomial:
        ols_fit_P = Polynomial(ols_fit_P, domain=[-1, 1])
    if fig is None:
        f, ax = plt.subplots(1,1, figsize=(8, 6))
    ax.plot(*oracle_P.linspace(), color="b", ls="-", lw=3, alpha=.7, label="$y=f(x)$")
    ax.plot(x_n, y_n, "ko", alpha=.7,label="$(y_i)_{i=1,\ldots,n}$")
    ax.plot(*ols_fit_P.linspace(), color="g", ls="-", lw=2, alpha=fit_alpha, label=fit_label)
    plt.legend(loc="lower right", prop={'size':20})
    return fig, ax

fig, ax = plot_polynomial_fit(w_oracle, w_ols, "$y=\hat{f}_{ols}(x)$")
```

### Forward stepwise algorithm:

- At step $i$, estimator $\mu$ and residuals $\mathbf{y} - \mu$

- Choose $\hat{j} = \underset{j}{\text{arg max}} \ | \hat{c_j}|$ where $\hat{c} = c(\hat{\mu}) = \mathbf{X}^T_j (\mathbf{y} - \mu)$

- $\hat{\mu} \rightarrow \hat{\mu} + \gamma \ \text{sign} (\hat{c_j}) \mathbf{x}_\hat{j}$

```python
##################
### Initialisation
##################
i = 0
# no intercept in x and mu because w_0 = 0
X_n_d = np.vstack((x_n, np.power(x_n,2), np.power(x_n,3))).T
m = d + 2
w_m_d = np.zeros((m, d))
mu_d = np.dot(X_n_d, w_m_d[i,:])
mu_d
```

```python
### Step i
w_stepwise = [0] + list(w_m_d[i,:])

### Step i -> i+1
print np.linalg.norm(y_n - np.dot(X_n_d, w_m_d[i,:]))
hat_c_d = np.dot(X_n_d.T, y_n - np.dot(X_n_d, w_m_d[i,:])) / [np.dot(x, x) for x in X_n_d.T]
hat_j = np.argmax(np.abs(hat_c_d))
print 'At step {}, hat_c_d = {} and hat_j = {}'.format(i, hat_c_d, hat_j)
w_m_d[i+1,:] = w_m_d[i,:]
w_m_d[i+1, hat_j] = hat_c_d[hat_j]
print 'w_d ({})= w_d ({}) + hat_c_d[{}] ({})'.format(w_m_d[i+1,:], w_m_d[i,:], hat_j, w_m_d[i+1,:] - w_m_d[i,:])


## plot
fig, ax = plot_polynomial_fit(w_oracle, [0] + list(w_m_d[i+1,:]), fit_alpha=.75, fit_label="$y=\hat{f}_{stepwise}(x)$")
for k in reversed(range(i)):
    fig, ax = plot_polynomial_fit(w_oracle, [0] + list(w_m_d[k,:]), fit_alpha=.5, fit_label="$y=\hat{f}_{stepwise}(x)$", fig=fig, ax=ax)

i += 1
```

## Bias-Variance trade-off

Mostly from that [course](http://www.youtube.com/watch?v=C3nIFH649wY&list=PLD0F06AA0D2E8FFBA&index=70)

- From a set of observed data $\mathcal{D}$, $\hat{\theta} = f(\mathcal{D})$ is an estimator of $\theta$. 
- Using the _squared-loss_ error function, the risk function $R(\hat{\theta};\theta, \mathcal{D}) = \mathbb{E}[(\hat{\theta} - \theta)^2 | \theta]$ can decomposed by :
    
$$ 
R(\hat{\theta};\theta, \mathcal{D}) = \underbrace{\mathbb{E}[(\hat{\theta} - \mathbb{E}[\hat{\theta}])^2]}_{\text{variance}} + \underbrace{(\mathbb{E}[\hat{\theta}] - \theta)^2}_{\text{bias}^2}
$$
